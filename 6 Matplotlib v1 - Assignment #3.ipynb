```python
# Exercise #1: Create a plot where y = x**2 for x over the interval 1 to 5, properly labelled
# Start here
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import matplotlib.mlab as mlab
import math

a=[]
b=[]
for x in range(0,6):
    y=x**2
    a.append(x)
    b.append(y)
    #x= x+1

fig= plt.figure()
axes=fig.add_subplot(111)
axes.plot(a,b)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend(("$y=x^2$",))
plt.show()
```

```python
# Exercise #2: Create a histogram where the mean = 0, std. dev. = 1, n = 300, and there are sqrt(n) bins
# The chart will be a little different each time you run it
import numpy
fig = plt.figure()
n=300

NUMBER_OF_BINS = int(numpy.sqrt(n))
# example data
mu = 0 # mean of distribution
sigma = 1 # standard deviation of distribution
x = mu + sigma * np.random.randn(n)

# the histogram of the data
n, bins, patches = plt.hist(x, NUMBER_OF_BINS, normed=1, facecolor='green', alpha=0.5)
# add a 'best fit' line
y = mlab.normpdf(bins, mu, sigma)
plt.plot(bins, y, 'r--')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Tweak spacing to prevent clipping of ylabel
plt.subplots_adjust(left=0.15)
```
